```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**2

x_values=np.linspace(-10, 10, 100)
y_values=f(x_values)
plt.plot(x_values, y_values , label="f(x)=x^2", color="blue")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Grafica de f(x)=x^2")
plt.legend()
plt.grid(True)
plt.show()
```
